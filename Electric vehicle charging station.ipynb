```python
import pandas as pd
import requests
```

```python
charging_station = pd.read_csv('./data/전기차 충전소 현황.txt', sep = "\t", engine='python', encoding = 'utf-8')
charging_station
```

```python
for i in charging_station.index:
    print(charging_station.loc[i, '주소'])
```

```python
url = 'https://dapi.kakao.com/v2/local/search/address.json'
headers = {'Authorization': 'KakaoAK '}    ## 이 부분 반드시 지우기!!!!!!!!!!!!!!!!!!
params = {'query':'서울 구로구 개봉동 477'}

# resp = requests.get(url, headers=headers, params=params).json()
# print(resp)

# cnt = len(charging_station.index)
```

```python
lngLat = []

for i in charging_station.index:
    #if i == 10:
    #    break
    params['query'] = charging_station.loc[i, '주소']
    print(i,charging_station.loc[i, '주소'])
    try:
        resp = requests.get(url, headers=headers, params=params).json()
        lngLat.append((resp['documents'][0]['x'], resp['documents'][0]['y']))
        print("SUCCEESS!    ", (resp['documents'][0]['x'], resp['documents'][0]['y']))
    except:
        lngLat.append((0,0))
        print("FAILE!")
```

```python
# print(lngLat)

for i in charging_station.index:
    if lngLat[i][0] == 0:
        print(charging_station.loc[i,'주소'])
```

```python
toCsv = pd.DataFrame(lngLat, columns = ['Lat', 'Lng'])
toCsv.to_csv('./data/충전소 좌표.csv')
```
