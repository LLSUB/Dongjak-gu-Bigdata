```python
import pandas as pd
```

```python
seoul_car_data = pd.read_csv('./data/서울기 자동차 등록 현황.txt', sep = "\t", engine='python', encoding = 'utf-8')
seoul_car_data
```

```python
seoul_car_data = seoul_car_data.fillna('0')
```

```python
total_car_num = {}

for r in seoul_car_data.index:
    gu = seoul_car_data.loc[r, '사용본거지법정동명'].split()
    print(gu, seoul_car_data.loc[r, '동별 총 대수'])
    try:
        num = int(seoul_car_data.loc[r, '동별 총 대수'].replace(',',''))
        if gu[1] in total_car_num.keys():
            total_car_num[gu[1]] += num
        else:
            total_car_num[gu[1]] = num
        print("Success")
    except:
        print("Last or Error")
```

```python
total_car_num.keys()
```

```python
i = 1

sorted_car_num = sorted(total_car_num.items(), key=lambda x : x[1], reverse = True)

for name, v in sorted_car_num:
    if name == '동작구':
        print("{:>4}{:>8}{:>10}    <<<".format(i,name,v))
    else:
        print("{:>4}{:>8}{:>10}".format(i,name,v))
    i += 1
```

```python
electric_car_num = {}

for r in seoul_car_data.index:
    gu = seoul_car_data.loc[r, '사용본거지법정동명'].split()
    print(gu, seoul_car_data.loc[r, '전기'])
    try:
        num = int(seoul_car_data.loc[r, '전기'].replace(',',''))
        if gu[1] in electric_car_num.keys():
            electric_car_num[gu[1]] += num
        else:
            electric_car_num[gu[1]] = num
        print("Success")
    except:
        print("Last or Error")
```

```python
i = 1

sorted_electric_car = sorted(electric_car_num.items(), key=lambda x : x[1], reverse = True)
print("<서울시 전기차 대수 정렬>\n")
for name, v in sorted_electric_car:
    if name == '동작구':
        print("{:>4}{:>8}{:>10}    <<<".format(i,name,v))
    else:
        print("{:>4}{:>8}{:>10}".format(i,name,v))
    i += 1
```

```python
ratio_electric = {}
i = 0

for k in total_car_num.keys():
    ratio_electric[k] = electric_car_num[k] / float(total_car_num[k])

S_ratio = sorted(ratio_electric.items(), key=lambda x: x[1], reverse = True)

print("{:>4}{:>10}{:>15}{:>10}{:>20}".format("순위", "구", "총 자동차 대수", "전기차 대수", "<비율>"))
print("-"*80)
for name, v in S_ratio:
    i += 1
    if name == '동작구':
        print("{:>4}{:>10}{:>15}{:>15}{:>30.4}    <<<".format(i, name, total_car_num[name], electric_car_num[name], v))
    else:
        print("{:>4}{:>10}{:>15}{:>15}{:>30.4}".format(i, name, total_car_num[name], electric_car_num[name], v))
```
